```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
goodreads_file = "Resources/goodreads_booksv2.csv"
goodreads_df = pd.read_csv(goodreads_file)
goodreads_df.head()

# https://stackoverflow.com/questions/37333299/splitting-a-column-by-delimiter-pandas-python
```

```python
goodreads_selected_df = pd.DataFrame()
goodreads_selected_df = goodreads_df[["isbn13","title","authors","publisher","publication_date","average_rating"]]
goodreads_selected_df.head()
```

```python
#Renaming Columns GoodReads:
goodreads_selected_df = goodreads_selected_df.rename(columns={'isbn13':'isbn',"authors":"author",'average_rating':'goodreads_rtg'})
goodreads_selected_df.head()
```

```python
google_file = "Resources/google_books.csv"
google_df = pd.read_csv(google_file)
google_df.head(3)
```

```python
google_selected_df = pd.DataFrame()
google_selected_df = google_df[["ISBN","rating"]]
google_selected_df.head()
```

```python
#Renaming Columns GoodReads:
google_selected_df = google_selected_df.rename(columns={'ISBN':'isbn','rating':'google_rtg'})
google_selected_df.head()
```

```python
# Merge this Goodreads with Google DF on ISBN
books = pd.merge(goodreads_selected_df, google_selected_df, on=["isbn"])
books.head()
```
